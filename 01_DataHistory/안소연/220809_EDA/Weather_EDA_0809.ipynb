### 도로 구분 없이 기상상태별로 합치기

```python
# -는 자료가 없다는 표시. 한 해의 자료가 통째로 없는게 아니라 그 부분만 없는걸 보면 일단 사고가 일어나지 않은거라 판단하여 0으로 대체
acc_group = df1.replace('-', 0)
acc_group
```

```python
#도로 열 필요없어서 지움
acc_group = acc_group.drop(['도로'], axis=1)
acc_group
```

```python
acc_group.iloc[:, 2:] = acc_group.iloc[:, 2:].astype('int')
```

```python
acc_group = acc_group.groupby(by=['기상', '월']).sum().groupby(level=[0]).cumsum()
acc_group
```

### 치사율 데이터

```python
#부상 열 필요없어서 지움
data = acc_group.drop(['부상_2017', '부상_2018', '부상_2019', '부상_2020', '부상_2021'], axis=1)
data
```

```python
#계산하기 쉽게 열 배열 바꾸기
data = data[['사고_2017', '사고_2018', '사고_2019', '사고_2020', '사고_2021','사망_2017', 
            '사망_2018', '사망_2019', '사망_2020','사망_2021']]
data
```

```python
#치사율(명/건)
fatality = (data.iloc[:,5:10].values / data.iloc[:,:5].values)*100
row = data.index

data_fatality = pd.DataFrame(fatality, index=row, columns=['2017','2018','2019','2020', '2021'])
data_fatality
```

```python
dd = data_fatality.reset_index(level=['월'])
dd
```

### 사망률 데이터

```python
#사고 열 필요없어서 지움
acc_data = acc_group.drop(['사고_2017', '사고_2018', '사고_2019', '사고_2020', '사고_2021'], axis=1)
acc_data
```

```python
#사상자(사망+부상)
df1 = pd.DataFrame(acc_data.iloc[:, 0:2].sum(axis=1), columns=['사망_2017'])
df2 = pd.DataFrame(acc_data.iloc[:, 2:4].sum(axis=1), columns=['사망_2018'])
df3 = pd.DataFrame(acc_data.iloc[:, 4:6].sum(axis=1), columns=['사망_2019'])
df4 = pd.DataFrame(acc_data.iloc[:, 6:8].sum(axis=1), columns=['사망_2020'])
df5 = pd.DataFrame(acc_data.iloc[:, 8:10].sum(axis=1), columns=['사망_2021'])
```

```python
df1
```

```python
#사망률(사망자/사상자)
death_2017 = (acc_data.iloc[:, 0:1]/df1)*100
death_2018 = (acc_data.iloc[:, 2:3]/df2)*100
death_2019 = (acc_data.iloc[:, 4:5]/df3)*100
death_2020 = (acc_data.iloc[:, 6:7]/df4)*100
death_2021 = (acc_data.iloc[:, 8:9]/df5)*100

death_2017
```

```python
death_toal = [death_2017, death_2018, death_2019, death_2020, death_2021]
```

```python
death = pd.concat(death_toal, axis=1)
death
```

```python
#컬럼명 변경
death.columns = ['2017','2018','2019','2020', '2021']
death
```

```python
#기상 인덱스만 살려놓기
d = death.reset_index(level=['월'])
d 
```

```python
#인덱스 모두 제거
dead = death.reset_index(level=['기상', '월'])
dead
```

```python
#행 열 바꾸기
dead1 = dead.transpose()
dead1
```
