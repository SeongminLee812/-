### 도로 구분 없이 기상별로 데이터 합치기

```python
accident = df.groupby(['기상']).sum()
accident
```

```python
#2017년 사고 데이터 기준으로 정렬
accident.sort_values(by='사고_2017', ascending=False, inplace = True)
accident
```

```python
#사고, 사망, 부상 순서로 열 다시 정렬(계산하기 쉽게하려고)
accident_group = accident[['사고_2017', '사고_2018', '사고_2019', 
                                 '사고_2020', '사고_2021','사망_2017', 
                                '사망_2018', '사망_2019', '사망_2020',
                                '사망_2021', '부상_2017', '부상_2018', '부상_2019', '부상_2020', '부상_2021']]
accident_group
```

#### 교통사고 날씨비율 데이터

```python
#최근5년치 사고건수
accident_total = pd.DataFrame(accident_group.iloc[:,:5])
accident_total
```

```python
accident_total.to_csv('C:/Users/ankm6/DS_Project/weather_5years.csv') #csv파일로 생성
```

```python
#사고건수 비율(100%기준)
accident_total = accident_total.div(accident_total.sum(axis=0), axis=1)*100
accident_total
```

```python
#컬럼명 변경
accident_total.columns = ['2017','2018','2019','2020', '2021']
accident_total
```

```python
#행열 자리 바꿈
accident_total1= accident_total.transpose()
accident_total1
```

#### 치사율 데이터

```python
fatality = (accident_group.iloc[:,5:10].values / accident_group.iloc[:,:5].values)*100
row = ['맑음', '비', '흐림', '눈', '기타/불명', '안개']

fatality_data = pd.DataFrame(fatality, index=row, columns=['2017','2018','2019','2020', '2021'])
fatality_data
```

```python
#행열 자리 바꿈
fatality_data = fatality_data.transpose()
fatality_data
```

#### 사망률 데이터

```python
accident
```

```python
#전체 사상자 수(사망자+부상자)
df1 = pd.DataFrame(accident.iloc[:, 1:3].sum(axis=1), columns=['사망_2017'])
df2 = pd.DataFrame(accident.iloc[:, 4:6].sum(axis=1), columns=['사망_2018'])
df3 = pd.DataFrame(accident.iloc[:, 7:9].sum(axis=1), columns=['사망_2019'])
df4 = pd.DataFrame(accident.iloc[:, 10:12].sum(axis=1), columns=['사망_2020'])
df5 = pd.DataFrame(accident.iloc[:, 13:15].sum(axis=1), columns=['사망_2021'])

df4
```

```python
#사망률(사망자/전체 사상자)
death_2017 = (accident.iloc[:, 1:2]/df1)*100
death_2018 = (accident.iloc[:, 4:5]/df2)*100
death_2019 = (accident.iloc[:, 7:8]/df3)*100
death_2020 = (accident.iloc[:, 10:11]/df4)*100
death_2021 = (accident.iloc[:, 13:14]/df5)*100

death_2021
```

```python
death_toal = [death_2017, death_2018, death_2019, death_2020, death_2021]
```

```python
#사망률 하나로 합치기
death = pd.concat(death_toal, axis=1)
death
```

```python
#컬럼명 변경
death.columns = ['2017','2018','2019','2020', '2021']
death
```

```python
#행열 자리 바꿈
death = death.transpose()
death
```

### 도로 구분 없이 기상별로 합치기

```python
# -는 자료가 없다는 표시인데 0으로 대체함
acc_group = df1.replace('-', 0)
acc_group
```

```python
#도로 열 필요없어서 지움
acc_group = acc_group.drop(['도로'], axis=1)
acc_group
```

```python
#타입변환
acc_group.iloc[:, 2:] = acc_group.iloc[:, 2:].astype('int')
```
